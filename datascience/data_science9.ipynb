# Exercise 5
Perform a dendrogram using the Tanimoto distance.

```python
rownames,colnames,data = readfile("data/zebo.txt")
```

```python
def tanamoto(v1,v2):
 c1,c2,shr=0,0,0

 for i in range(len(v1)):
    if v1[i]!=0: c1+=1 # in v1
    if v2[i]!=0: c2+=1 # in v2
    if v1[i]!=0 and v2[i]!=0: shr+=1 # in both

 return 1.0-(float(shr)/(c1+c2-shr))
```

```python
clust = hcluster(data, distance=tanamoto)
```

```python
img = drawdendrogram(clust,labels=rownames)
```

```python
display(img)
```
