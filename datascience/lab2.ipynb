```r
library(ggplot2)

data <- read.csv("data/Queratocono.csv")
head(data)
```

```r
ggplot(data, aes(x = K1, y = K2)) +
  geom_point() +
  geom_smooth()

ggplot(data, aes(x = K1, y = K2)) +
  geom_point() +
  geom_smooth(method="lm")
```

```r
ggplot(data, aes(x = K1, y = K2, color = factor(na)))+
     geom_point(shape = 18, size=3) +
     geom_smooth(method="lm") +
     labs(title = "Relation between K1 and K2") +
     scale_color_discrete(name = "factor (na)")
```

```r
ggplot(data, aes(x = K1, y = K1.salida))+
    geom_point() +
    scale_color_discrete(name = "factor (na)")
```

```r
ggplot(data, aes(x = factor(grosor), fill = factor(na))) +
  geom_bar()
```

```r
ggplot(data, aes(x = K1, y = K2, color = factor(grosor))) +
  geom_point(alpha = 1/3) +
  facet_grid(diam ~ factor(na))
```

```r
# install.packages("gridExtra")
library(gridExtra)

options(
  repr.plot.width = 12,  # Adjust the width as needed
  repr.plot.height = 6  # Adjust the height as needed
)


plot_k1 <- ggplot(data, aes(x = factor(grosor), y = K1)) +
  geom_boxplot()

plot_k2 <- ggplot(data, aes(x = factor(grosor), y = K2)) +
  geom_boxplot()

grid.arrange(plot_k1, plot_k2, ncol = 2)
```
